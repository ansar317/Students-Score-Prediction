# Students Score

# Import Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
Students_Score_Dataset = pd.read_csv('student_scores.csv')
Students_Score_Dataset.head()
```

# Linear Regression Model

```python
reg = LinearRegression()
x = Students_Score_Dataset[['Hours']]
y = Students_Score_Dataset['Scores']
reg.fit(x,y)
```

# Prediction

```python
# Get the value of Hours from the user
custom_hours = float(input("Enter hours of study: "))

# Predict score based on user input
predicted_score = reg.predict([[custom_hours]])

# Display the predicted score
print(f"Predicted Score for {custom_hours} Hours of Study: {predicted_score[0]}")

# Visualize the Results
plt.scatter(x, y, label='Actual Data')
plt.plot(x, reg.predict(x), color='red', linewidth=3, label='Linear Regression')
plt.scatter(custom_hours, predicted_score, color='green', label='Predicted Value', marker='*', s=100)
plt.title("Linear Regression: Hours vs. Predicted Score")
plt.xlabel("Hours of Study")
plt.ylabel("Predicted Score")
plt.legend()
plt.show()
```
